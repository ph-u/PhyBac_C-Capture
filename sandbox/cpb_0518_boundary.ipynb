# sensitivity analysis

```r
dEfIni = rep(1e-5,3)
dEfpAr = c(.5,.875,.63,.3,.2,.6,.55,5,.14)
dEfpAN = c("x","e_PR","e_P","g_P","a_P","e_BR","e_B","g_B","m_B")
ebcAlt(dEfpAr)
```

```r
pApLt = function(pA, rAg=c(0,5), eT=100, iniPop=dEfIni, name=dEfpAN, values=dEfpAr){
    a0 = as.data.frame(matrix(NA,nc=5,nr=eT))
    sEq=seq(rAg[1],rAg[2],(rAg[2]-rAg[1])/eT)
    for(i in 1:eT){
        tpAr = values
        tpAr[pA] = sEq[i]
        a = ebcData(iniPop = iniPop, parameter = tpAr)
        a0[i,] = c(sEq[i],a[nrow(a),-1])
    };rm(i)
    colnames(a0) = c(name[pA],"C","P","B","A")

    matplot(a0[,1],a0[,-1], type="l", lty=1, lwd=5, col=cBp[-c(3,5)], xlab=colnames(a0)[1], ylab="carbon density (gC/m^3)", cex.lab=2, cex.axis=2, cex.main=2, main=paste0(name[-pA]," = ",tpAr[-pA], collapse=", "))
    legend("topleft", inset=c(0,0), legend = paste0(colnames(a0)[-1],": ",round(a0[nrow(a0),-1],2)), pch = rep(16,3), col = cBp[-c(3,5)], bty="n", cex = 2)
}
```

# boundary initial population

```r
for(i in 1:length(cAtid)){if(i!=1 & i!=10){
    assign(paste0("cAt",cAtid[i]),cAt[which(cAt$eqmid==cAtid[i]),8:16])
}};rm(i)
```

```r
bOndary = function(df, row){
    pA = as.numeric(df[row,])
    name = colnames(df[row,])
    
    a1 = ebcAlt(pA,2)
    a1 = unique(a1[which(a1!=Inf)])
    a1 = a1[order(a1)]

    a = as.data.frame(matrix(NA,nc=5,nr=13))
    for(i in 1:13){
        a0 = ebcData(iniPop=10^-(i-1),parameter = pA)
        a[i,] = c(-(i-1), a0[nrow(a0),-1])
    };rm(i)
    colnames(a) = c("log10(iniPop)","C","P","B","A")

    matplot(a[,1],a[,-1], type="l", lty=1, lwd=5, col=cBp[-c(3,5)], xlab=colnames(a)[1], ylab="carbon density (gC/m^3)", cex.lab=2, cex.axis=2, cex.main=2, main=paste0(name," = ",pA, collapse=", "), ylim=c(min(a1),(max(a1)*1.1)))
    legend("topright", inset=c(0,0), legend = paste0(colnames(a)[-1],": ",signif(a[nrow(a),-1],2)), pch = rep(16,3), col = cBp[-c(3,5)], bty="n", cex = 2)
}
```
