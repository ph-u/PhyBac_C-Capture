# given equilibria positions (from SymPy)

Sol | ....... C ....... | ....... P ....... | ..................... B .....................
--- | --- | --- | ---
(1) | 0 | 0 | 0
(2) | $\frac{m_B}{e_{BR}e_B g_B}$ | 0 | $\frac{x}{g_B(e_{BR} - 1)}$
(3) | $\frac{e_P(e_{PR}g_P)^2}{a_P x}$ | $\frac{e_{PR}e_Pg_P}{a_P}$ | 0
(4) | $\frac{m_B}{e_{BR}e_B g_B}$ | $\frac{e_{PR}e_P g_P}{a_P}$ | $\frac{a_P m_B x - e_{BR}e_B g_B e_P(e_{PR}g_P)^2}{a_P m_B g_B(e_{BR}-1)}$

result expectation tip: only solution (3) & (4) is possible, and only solution 4 can create crashed system during integration

## plotting custom function

```r
ebcPlt = function(iniPop, pA){
    pA = as.numeric(pA[1:9])
    
    dAta=ebcData(endTime=1e3, iniPop=iniPop, parameter=pA)
    eQm=ebcEqm(pA)
    eQm=c(eQm,sum(eQm))
    
    par(mar=c(5, 5, 3, 1), xpd=TRUE, cex = 1) ## c(B,L,T,R)

    matplot(dAta[,1], dAta[,-1], type="l", lty=1, lwd=5, col=cBp[-c(3,5)], xlab=paste0(colnames(dAta)[1]," (day)"), ylab="carbon density (gC/m^3)", cex.lab=2, cex.axis=2, cex.main=2, main=paste0(colnames(sAm)[1:9]," = ",sAm0[1:9], collapse=", "))
    points(x=rep(dAta[nrow(dAta),1],ncol(dAta)-1), y=eQm, pch=rep(16,ncol(dAta)-1), col = cBp[-c(3,5)], cex=2)
    legend("topleft", inset=c(0,0), legend = paste0(colnames(dAta)[-1],": ",round(dAta[nrow(dAta),-1],2)), pch = rep(16,3), col = cBp[-c(3,5)], bty="n", cex = 2)
}
```

## initial population-independence & alternate eqm demo

```r
dAta=ebcData(endTime=1e3, iniPop=1e-12, parameter=sAm0[1:9])
eQm=ebcEqm(sAm0)
eQm=c(eQm,sum(eQm))

par(mfrow=c(1,2), mar=c(5, 5, 3, 1), xpd=TRUE, cex = 1) ## c(B,L,T,R)

matplot(dAta[,1], dAta[,-1], type="l", lty=1, lwd=5, col=cBp[-c(3,5)], xlab=paste0(colnames(dAta)[1]," (day)"), ylab="carbon density (gC/m^3)", cex.lab=2, cex.axis=2, cex.main=2)
points(x=rep(1000,ncol(dAta)-1), y=eQm, pch=rep(16,ncol(dAta)-1), col = cBp[-c(3,5)], cex=2)
legend("topleft", inset=c(0,0), legend = paste0(colnames(dAta)[-1],": ",signif(dAta[nrow(dAta),-1],3)), pch = rep(16,3), col = cBp[-c(3,5)], bty="n", cex = 2)

matplot(dAta[,1], dAta[,-1], type="l", lty=1, lwd=5, col=cBp[-c(3,5)], xlab=paste0(colnames(dAta)[1]," (day)"), ylab="carbon density (gC/m^3)", cex.lab=2, cex.axis=2, cex.main=2, ylim=c(-.0001,.0001), xlim=c(0,200))
points(x=rep(1000,ncol(dAta)-1), y=eQm, pch=rep(16,ncol(dAta)-1), col = cBp[-c(3,5)], cex=2)
```

```r
ebcPlt(1e-8, sAm) ## initial population, parameter set
```

```r
ebcPlt(1e-5, sAm) ## initial population, parameter set
```
