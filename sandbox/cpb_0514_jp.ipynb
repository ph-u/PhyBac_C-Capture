# given equilibria positions (from SymPy)

Sol | ....... C ....... | ....... P ....... | ..................... B .....................
--- | --- | --- | ---
(1) | 0 | 0 | 0
(2) | $\frac{m_B}{e_{BR}e_B g_B}$ | 0 | $\frac{x}{g_B(e_{BR} - 1)}$
(3) | $\frac{e_P(e_{PR}g_P)^2}{a_P x}$ | $\frac{e_{PR}e_Pg_P}{a_P}$ | 0
(4) | $\frac{m_B}{e_{BR}e_B g_B}$ | $\frac{e_{PR}e_P g_P}{a_P}$ | $\frac{a_P m_B x - e_{BR}e_B g_B e_P(e_{PR}g_P)^2}{a_P m_B g_B(e_{BR}-1)}$

result expectation tip: only solution (3) & (4) is possible, and only solution 4 can create crashed system during integration

## recap SymPy analytical solutions

```julia
## env set-up
@vars C P B
x, e_PR,e_P,g_P,a_P, e_BR,e_B,g_B,m_B = symbols("x e_PR e_P g_P a_P e_BR e_B g_B m_B",positive=true)

## equations
dC = g_P*e_PR*(1-e_P)*P +a_P*P^2 +g_B*(e_BR*(1-e_B)-1)*C*B +m_B*B -x*C
dP = g_P*e_PR*e_P*P -a_P*P^2
dB = g_B*e_BR*e_B*C*B -m_B*B

## analytical solving
solve([Eq(dC,0),Eq(dP,0),Eq(dB,0)],[C,P,B])
```

## plotting custom function

```julia
function ebcPlt(iniPop,para)
    dAta=ebcData(1000, iniPop,para)
    eQm=ebcEqm(para)
    eQm=push!(eQm,sum(eQm))
    plot(dAta, label=["C"*string(round(dAta[size(dAta)[1],1],digits=2)) "P"*string(round(dAta[size(dAta)[1],2],digits=2)) "B"*string(round(dAta[size(dAta)[1],3],digits=2)) "A"*string(round(dAta[size(dAta)[1],4],digits=2))], legend=:bottomleft)
    scatter!([1000,1000,1000,1000],eQm, label="analytical Sol")
end
```

```julia
function ebcRPlt(para)
    dAta=ebcData(1000,1e-5,para)
    rAte=DataFrame(C=Float16[],P=Float16[],B=Float16[],A=Float16[])
    for tmp in 1:size(dAta)[1]
        a = ebcRate(dAta[tmp,1:3],para)
        a = push!(a,sum(a))
        push!(rAte,a)
    end
    plot(Array(rAte),label=["C: "*string(rAte[size(rAte)[1],1]) "P: "*string(rAte[size(rAte)[1],2]) "B: "*string(rAte[size(rAte)[1],3]) "A: "*string(rAte[size(rAte)[1],4])])
end
```
